```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset from sklearn
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target (species) column to the dataframe
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display the first few rows of the dataset
df.head()
```

```python
# Get info about the data types and missing values
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Basic statistics of numerical columns
df.describe()
```

```python
# Group by species and compute the mean for each group
df.groupby('species').mean()
```

```python
# Create a line chart showing trends over samples for Sepal and Petal length
df['index'] = range(len(df))  # Add an index column for plotting
plt.plot(df['index'], df['sepal length (cm)'], label='Sepal Length')
plt.plot(df['index'], df['petal length (cm)'], label='Petal Length')
plt.title('Sepal and Petal Length over Samples')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a bar chart showing average Petal Length per species
df.groupby('species')['petal length (cm)'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Petal Length per Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Create a histogram of Sepal Width
plt.hist(df['sepal width (cm)'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Sepal Width Distribution')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot for Sepal Length vs Petal Length
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Sepal vs Petal Length by Species')
plt.show()
```

```python
### Findings:
- Setosa species has the shortest petals.
- Virginica species has generally larger petal lengths and sepal lengths.
- The relationship between petal and sepal length is positive, with larger petals corresponding to larger sepals.
```
