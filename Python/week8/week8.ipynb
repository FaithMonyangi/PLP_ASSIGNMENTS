```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Keep only countries you're interested in
countries = ['Kenya', 'India', 'United States']
covid_df = df[df['location'].isin(countries)].copy()  # also added .copy() to be safe

# Convert 'date' to datetime safely
covid_df.loc[:, 'date'] = pd.to_datetime(covid_df['date'])

# Drop rows where total_cases is missing
covid_df = covid_df.dropna(subset=['total_cases'])

# Preview the cleaned DataFrame
covid_df.head()
```

```python
# Compare total cases over time
plt.figure(figsize=(12, 6))
for country in countries:
    subset = covid_df[covid_df['location'] == country]
    plt.plot(subset['date'], subset['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for country in countries:
    subset = covid_df[covid_df['location'] == country]
    plt.plot(subset['date'], subset['total_vaccinations'], label=country)

plt.title("Total Vaccinations Over Time")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
latest = covid_df.groupby('location').last()
latest['death_rate'] = latest['total_deaths'] / latest['total_cases']

sns.barplot(x=latest.index, y=latest['death_rate'])
plt.title("COVID-19 Death Rate by Country")
plt.ylabel("Death Rate")
plt.xlabel("Country")
plt.tight_layout()
plt.show()
```

```python
### ðŸ“Œ Findings
- The United States had the highest total cases and vaccinations.
- Kenya shows a lower death rate compared to India.
- All countries follow a similar upward trend in total cases and vaccinations over time.
```
